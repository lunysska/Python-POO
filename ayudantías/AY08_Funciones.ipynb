## Programando el juego de Piedra, Papel o Tijera

```python
from random import choice

while True:
  opcion=input("Vamos a jugar, escoje entre piedra,papel o tijera")
  compu=['piedra','papel','tijera']
  x=choice(compu)

  print("La opción de la computadora es: ", x)
  if opcion == 'piedra':
    if x == 'piedra':
      print("Empate")
    elif x == 'tijera':
      print("Ganaste")
    else:
      print("Perdiste")
    break
  elif opcion == 'tijera':
    if x == 'piedra':
      print("Perdiste")
    elif x == 'tijera':
      print("Empate")
    else:
      print("Ganaste")
    break
  else:
    if x == 'piedra':
      print("Ganaste")
    elif x == 'tijera':
      print("Perdiste")
    else:
      print("Empate")
    break
```

```python
# función con múltiples salidas
lista = [10,50,30,12,6,8,100]
def efunc(b):
    cotaMax = max(b)
    CotaMin = min(b)
    Primer = b[0]
    Ultimo = b[-1]
    return cotaMax,CotaMin,Primer,Ultimo

efunc(lista)
```

Una función compuesta, indica la función $f$ compuesta de una función $g$, y sigue la regla de composición $(f\circ g)(x)=f(g(x))$.

Si $f(x)=3x-1$ y $g(x)=x^3+2$, evaluar $(f\circ g)(10)=f(g(10))$:

```python
def f(x):
  y = 3*x - 1
  return y

def g(x):
  y = x**3 +2
  return y

def fog(f,g,x):
  y = f(g(x))
  return y

print(fog(f,g,10))
```

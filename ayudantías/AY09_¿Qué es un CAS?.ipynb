¿Todo en matemáticas puede verse como expresión?

```python
class Expresion:
    pass
```

Expresion no tiene atributos porque no sabemos todavía qué tipo de expresión será.

```python
class Numero(Expresion):
    def __init__(self, valor):
        self.valor = valor
```

```python
class Variable(Expresion):
    def __init__(self, nombre):
        self.nombre = nombre
```

```python
x = Variable("x")
dos = Numero(2)
```

```python
class Potencia(Expresion):
    def __init__(self, base, exponente):
        self.base = base
        self.exponente = exponente
```

```python
expr = Potencia(x, dos)
```

# Diccionarios

Los diccionarios se utilizan para almacenar valores de datos en pares clave:valor.

Un diccionario es una colección que está ordenada, cambiable y no permite duplicados.

Un diccionario de Python es una asignación de claves únicas a valores. Los diccionarios son mutables, lo que significa que se pueden cambiar. Un diccionario asigna un conjunto de objetos (claves/keys) a otro conjunto de objetos (valores).

```python
peso={} #diccionario vacio
```

```python
peso=dict() #diccionario vacio
```

```python
estudiante = {
    "nombre": "Jonathan",
    "edad": 24,
    "carrera": "Física"
}
```

```python
print(estudiante["nombre"])
```

Las llaves deben ser inmutables (números, strings, tuplas).

Los diccionarios no pueden tener dos elementos con la misma clave:

```python
estudiante = {
    "nombre": "Jonathan",
    "edad": 24,
    "carrera": "Física",
    "carrera": "Actuaría"
}
```

```python
print(estudiante)
```

Los valores duplicados sobrescribirán los valores existentes:

Los valores en los elementos del diccionario pueden ser de cualquier tipo de datos

```python
dict3 = dict(nombre = "Silvia", age = 24, country = "México")
print(dict3)
```

Programa que solicite al usuario el número de materias que lleva en el semestre, para posteriormente solicitar el nombre de cada materia, y que el usuario pueda ingresar su calificación.
El programa debe deveolver el promedio.

## Ejercicio:

```python
materias={}
lista=[]
n = int(input("¿Cuantas materias llevas en el semestre "))
i = 0
while i < n:
  i += 1
  lista.append(input("Nombre de la materia: "))

s= 0
for i in lista:
  materias[i] = float(input(f"Dame la calificación de {i}: "))
  s+= materias[i]

print("Tu promedio es de ", s/n)

```

```python
print(materias)
```
